# Extracting the Mean, Median & Mode:

```python
# MAHINDRA
```

```python
mahindra=t1a[(t1a.Make=='Mahindra')]
```

```python
mahindra3=mahindra.loc[0::,['Displacement',
                            'Fuel_Tank_Capacity_litre',
                            'City_Mileage_km_litre',
                            'Highway_Mileage_km_litre',
                            'Minimum_Turning_Radius']]

mahindra3.head()
```

```python
Mean=[]
Median=[]
SD=[]
CV=[]


for col in mahindra3:
    col_mean=mahindra3.loc[:,col].mean() 
    Median.append(mahindra3.loc[:,col].median()) 
    Mean.append(col_mean) 
    col_sd=mahindra3.loc[:,col].std() 
    SD.append(col_sd) 
    CV.append(col_sd/col_mean*100) 
    
    
row_head=['Median',
          'Mean',
          'Standard Deviation',
          'Coefficient of Variation']

col_names=['Displacement',
           'Fuel_Tank_Capacity_litre',
           'City_Mileage_km_litre',
           'Highway_Mileage_km_litre',
           'Minimum_Turning_Radius']

x_data=pd.DataFrame([Median,Mean, SD, CV],columns=col_names)
x_data.insert    (0, "Mahindra", row_head, True)
x_data
```

# TASK 2

```python
t2=data.groupby(by='Make')['Displacement',
                           'Fuel_Tank_Capacity_litre',
                           'City_Mileage_km_litre',
                           'Highway_Mileage_km_litre',
                           'Minimum_Turning_Radius'].mean()

t2.insert(5,'avgmileage',np.nan)
t2.head()
```

```python
t2.avgmileage=(t2.City_Mileage_km_litre+t2.Highway_Mileage_km_litre)/2
```

```python
t2.dropna(subset=['Displacement',
                  'Fuel_Tank_Capacity_litre',
                  'City_Mileage_km_litre',
                  'Highway_Mileage_km_litre',
                  'Minimum_Turning_Radius'],axis=0, inplace=True)

t2.head()
```

```python
plt.scatter(t2['avgmileage'],t2['Fuel_Tank_Capacity_litre'],color='magenta')
plt.title('Mileage vs. fuel tank capacity')
plt.xlim(5,25)
plt.ylim(30,100)
```

```python
plt.scatter(t2['avgmileage'],t2['Displacement'],color='green')
plt.title('mileage vs displacement')
```

```python
plt.scatter(t2['avgmileage'],t2['Minimum_Turning_Radius'],color='black')
plt.title('mileage vs turning radius')
```

# CORRELATION:

```python
corr1=t2['avgmileage'].corr(t2['Fuel_Tank_Capacity_litre'])
print('correlation for corr1 is ',round(corr1,3))
```

```python
corr2=t2['avgmileage'].corr(t2['Displacement'])
print('correlation for corr2 is ',round(corr2,3))
```

```python
corr3=t2['avgmileage'].corr(t2['Minimum_Turning_Radius'])
print('correlation for corr3 is ',round(corr3,3))
```
