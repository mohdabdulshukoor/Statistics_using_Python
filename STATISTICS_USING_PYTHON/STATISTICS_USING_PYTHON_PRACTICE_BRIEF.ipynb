# 3.1.4 Stratified  Sampling with respect to year and genre

```python
grouped_gs_df=gamesales_df.groupby(['Year'])
h=gamesales_df.Year.unique()
#g=gamesales_df.Genre.unique()
gs_Sample_df=pd.DataFrame()    #creating an empty DataFrame


for i in h:
    
    gs_data=grouped_gs_df.get_group(i)
    
    sample_df=gs_data.sample(n=30,
                             replace=True,
                             random_state=1) 
    
    gs_Sample_df=gs_Sample_df.append(sample_df,
                                     ignore_index=True)#Appending in the empty DataFrame which created before.
    
gs_Sample_df.head()
```

```python
grouped_gs_df=gamesales_df.groupby(['Genre'])
h=gamesales_df.Genre.unique()
#g=gamesales_df.Year.unique()
gs_Sample_df=pd.DataFrame()    #creating an empty DataFrame


for i in h:
    
    gs_data=grouped_gs_df.get_group(i)
    
    sample_df=gs_data.sample(n=30,
                             replace=True,
                             random_state=1) 
    
    gs_Sample_df=gs_Sample_df.append(sample_df,
                                     ignore_index=True)#Appending in the empty DataFrame which created before.
    
gs_Sample_df.head()
```

# 3.1.5 Clustered Sampling with respect to year and genre

```python
grouped_gy_df=gamesales_df.groupby(['Year'])
yrs=[2019,2020]
gy_Sample_df=pd.DataFrame()

for i in yrs:
    
    gy_data=grouped_gy_df.get_group(i)
    
    sample_gy_data=gy_data.sample(frac=.3,
                                    random_state=1)
    
    gy_Sample_df=gy_Sample_df.append(sample_gy_data,
                                     ignore_index=True)
    
gy_Sample_df.head()
```

```python
grouped_gy_df=gamesales_df.groupby(['Genre'])
yrs=['Action','Sports']
gy_Sample_df=pd.DataFrame()

for i in yrs:
    
    gy_data=grouped_gy_df.get_group(i)
    
    sample_gy_data=gy_data.sample(frac=.3,
                                    random_state=1)
    
    gy_Sample_df=gy_Sample_df.append(sample_gy_data,
                                     ignore_index=True)
    
gy_Sample_df.head()
```

```python
#Sampling:
#sample_data10=car_df.sample(n=10, replace=True,random_state=1)
mm1=gamesales_df.sample(n=10, #sample10
             replace=True,random_state=1)

mm2=gamesales_df.sample(n=20, #sample20
                        random_state=1)

mm3=gamesales_df.sample(n=30, #sample30
                        random_state=1)

mm4=gamesales_df.sample(n=40, #sample40
                        random_state=1)
```

```python
#Mean, Median, standard Deviation:

mm1a=st.mean(mm1['Global_Sales_M$'])
mm1b=st.stdev(mm1['Global_Sales_M$'])
mm1c=st.median(mm1['Global_Sales_M$'])
print("Sample-1 Mean= ",mm1a)
print("Sample-1 Median= ",mm1b)
print("Sample-1 Standard Deviation= ",mm1c)

mm2a=st.mean(mm2['Global_Sales_M$'])
mm2b=st.stdev(mm2['Global_Sales_M$'])
mm2c=st.median(mm2['Global_Sales_M$'])
print("Sample-2 Mean= ",mm2a)
print("Sample-2 Median= ",mm2b)
print("Sample-2 Standard Deviation= ",mm2c)
#
mm3a=st.mean(mm3['Global_Sales_M$'])
mm3b=st.stdev(mm3['Global_Sales_M$'])
mm3c=st.median(mm3['Global_Sales_M$'])
print("Sample-3 Mean= ",mm3a)
print("Sample-3 Median= ",mm3b)
print("Sample-3 Standard Deviation= ",mm3c)

mm4a=st.mean(mm4['Global_Sales_M$'])
mm4b=st.stdev(mm4['Global_Sales_M$'])
mm4c=st.median(mm4['Global_Sales_M$'])
print("Sample-4 Mean= ",mm4a)
print("Sample-4 Median= ",mm4b)
print("Sample-4 Standard Deviation= ",mm4c)
```
