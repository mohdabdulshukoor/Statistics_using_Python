## Demo - 3.2: Simple Random Sampling

### Simple Random Sampling without Replacement

```python
emp_df.sample(n=20, random_state=1)
```

### Simple Random Sampling with Replacement

```python
emp_df.sample(n=20, replace=True, random_state=1)
```

## Demo - 3.3: Systematic Sampling

```python
import numpy as np

indexes=np.arange(0,len(emp_df), step=5)     #( START, STOP, STEP)

print(emp_df.iloc[indexes])
```

## Demo - 3.4: Stratified Sampling

```python
grouped_d_df=emp_df.groupby(['Department'])

departments=emp_df.Department.unique()

Strat_Sample_df=pd.DataFrame()    #creating an empty DataFrame

for d in departments:
    d_data=grouped_d_df.get_group(d)
    sample_data=d_data.sample(n=5,replace=True, random_state=1)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index=True)#Appending in the empty DataFrame which created before

Strat_Sample_df
```

## Demo - 3.5: Cluster Sampling

```python
grouped_dept_df=emp_df.groupby(['Department'])
depts=['HR','IT']
C_Sample_df=pd.DataFrame()

for dept in depts:
    dept_data=grouped_dept_df.get_group(dept)
    sample_data=dept_data.sample(frac=.2, random_state=1)
    C_Sample_df=C_Sample_df.append(sample_data, ignore_index=True)
    
C_Sample_df
```

## Demo - 3.8: Distribution of Data

### Draw the shape of the distribution of the whole dataset

```python
import matplotlib.pyplot as plt

data_p=[54, 55, 59, 63, 64, 68, 69, 70]

plt.hist(data_p,bins=4)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
```

### Calculate distribution Characteristics

```python
import statistics as st
import scipy.stats as sts

print("Size ="+str(len(data_p)))
print("Mean ="+str(st.mean(data_p)))
print("Median ="+str(st.median(data_p)))
print("Standard Deviation ="+str(st.stdev(data_p)))
print("Shewness ="+str(sts.skew(data_p,bias=False)))
print("Kurtosis ="+str(sts.kurtosis(data_p,bias=False)))        #False
```

### Present the shape of distribution of horsepower(hp) of all cars present in the dataset

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(car_df['hp'], bins=5)
plt.xlabel("HP")
plt.ylabel("Frequency")
```

### Draw the random samples of sizes of 10, 20, and 30

```python
sample_data10=car_df.sample(n=10,
                            replace=True,
                            random_state=1)

sample_data20=car_df.sample(n=20,
                            replace=True,
                            random_state=1)

sample_data30=car_df.sample(n=30,
                            replace=True,
                            random_state=1)

print(sample_data10)
print(sample_data20)
print(sample_data30)
```

### Present the shape of distribution of each sample

```python
plt.hist(sample_data10['hp'], bins=5)
plt.xlabel("HP")
plt.ylabel("Frequency")

```

```python
plt.hist(sample_data30['hp'], bins=5)
plt.xlabel("HP")
plt.ylabel("Frequency")
```

### Calculate mean, median, mode, standard deviation, skewness and kurtosis of population and all sample datasets

```python
import statistics as st
import scipy.stats as sts
```

```python
p_mean=st.mean(car_df['hp'])
p_median=st.median(car_df['hp'])
p_mode=st.mode(car_df['hp'])
p_sd=st.stdev(car_df['hp'])

print("Population Mean= "+str(p_mean))
print("Population Median= "+str(p_median))
print("Population Mode= "+str(p_mode))
print("Population SD= "+str(p_sd))
print("Shewness ="+str(sts.skew(car_df['hp'],bias=False)))
print("Kurtosis ="+str(sts.kurtosis(car_df['hp'],bias=False)))
```
