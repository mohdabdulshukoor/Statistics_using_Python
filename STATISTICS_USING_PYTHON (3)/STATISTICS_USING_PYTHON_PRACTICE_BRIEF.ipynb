# TASK 1

# 1.1

```python
g_X=2000
μ=1332   #Average given!
σ=725

#calculating the z-score with the given mean value, X and standard deviation
 
z_score1=(g_X-μ)/σ
print(z_score1)
```

```python
lowerlimit=μ-4*σ
upperlimit=μ+4*σ

print("lowerlimit:",lowerlimit)
print("upperlimit:",upperlimit)

nlu=np.arange(lowerlimit,upperlimit)

plt.plot(nlu, norm.pdf(nlu, μ, σ))   #ploting
plt.title("Tax Distribution Curve")
plt.xlabel("Tax Returns")
plt.ylabel("Probability")
```

```python
print("Prospect of tax lesser than 2000:",norm.cdf(z_score1))
print("Prospect of tax greater than 2000:",(1-norm.cdf(z_score1)))

#"pvalue" is norm.cdf(z_score)
```

```python
x=0
μ=1332
σ=725

z_score2=(x-μ)/σ
#Probability of tax, which the proportion of tax indicates that a taxpayer owes to the government! 
print("Prospect of the tax (Zzz-score)",z_score2)

p_value=norm.cdf(z_score2)
print("p-value: ",p_value)
print("proportion: ",p_value*100)
```
