## Demo - 5.3: Normal Probability Distribution

### Read data from source file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from scipy.stats import norm
```

```python
#Reading the csv file using 'pandas'!

emp_df=pd.read_csv("DS1_C5_S5_Continuous_Concept_Employee_Data.csv")
emp_df[:5]
```

### Extract sample data using stratified sampling techniques. (Make sure that the sample size must be greater than 30)

```python
#sampling data greater than 30 ["stratified sampling"]

sample_emp_df=emp_df.groupby(['Department'])

departments=emp_df.Department.unique()

Strat_Sample_df=pd.DataFrame()

for d in departments:
    
    d_data=sample_emp_df.get_group(d)

    sample_data=d_data.sample(n=10,
                          replace=True,
                          random_state=1)

    Strat_Sample_df=Strat_Sample_df.append(sample_data, 
                                       ignore_index=True)

Strat_Sample_df
```

### Central Tendency of Sample

```python
import statistics as st
import scipy.stats as sts
```

```python
#mean & median of sample:

s_mean=st.mean(Strat_Sample_df['Annual_Salary($)'])
s_median=st.median(Strat_Sample_df['Annual_Salary($)'])

print(s_mean)
print(s_median)
```

### Central Tendency of Population

```python
#mean & median of population:

p_mean=st.mean(emp_df['Annual_Salary($)'])
p_median=st.median(emp_df['Annual_Salary($)'])

print(p_mean)
print(p_median)
```

### Standard Deviation of population and sample

```python
#

s_std=st.stdev(Strat_Sample_df['Annual_Salary($)'])
print(s_std)

#

p_std=st.stdev(emp_df['Annual_Salary($)'])
print(p_std)
```

### Probability Distribution curve of population and sample

```python
lower_limit=s_mean-4*s_std
upper_limit=s_mean+4*s_std

print(lower_limit)
print(upper_limit)
```

```python
x=np.arange(lower_limit,upper_limit)
plt.plot(x,norm.pdf(x,p_mean,p_std))
```

```python
xx=np.arange(lower_limit,upper_limit)
plt.plot(x,norm.pdf(x,s_mean,s_std))
```

## Demo - 5.5: Standard Normal Distribution

### Read data from source file

```python
import pandas as pd
```

```python
emp_df=pd.read_csv('DS1_C5_S5_Continuous_Concept_Employee_Data.csv')
print(type(emp_df))
print(emp_df.head())
print(emp_df.tail())
```

### Extract sample data using stratified sampling techniques. (Make sure that the sample size must be 40)

```python
grouped_g_df=emp_df.groupby(['Gender'])
genders=emp_df.Gender.unique()
Strat_Sample_df=pd.DataFrame()

for g in genders:
    g_data=grouped_g_df.get_group(g)
    sample_data=g_data.sample(n=40, random_state=1, replace=True)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index=True)
print(Strat_Sample_df)
```

### Calculate the population mean and the standard deviation

```python
import statistics as st
data_p=emp_df['Annual_Salary($)'].tolist()
p_mean = st.mean(data_p)
print("Population Mean ="+str(p_mean))

p_sd = st.stdev(data_p)
print("Population SD ="+str(p_sd))
```

### Calculate the samples mean and standard deviation

```python
data_s=Strat_Sample_df['Annual_Salary($)'].tolist()
s_mean = st.mean(data_s)
print("Sample Mean ="+str(s_mean))
s_sd = st.stdev(data_s)
print("Sample SD ="+str(s_sd))
lower=s_mean-4*s_sd
upper=s_mean+4*s_sd
```

### Calculate z-scores

```python
sample_mean=80000
z_score=(sample_mean-s_mean)/s_sd
print("Z-score= "+str(z_score))
```

### Calculate probability values of the z-score

```python
from scipy.stats import norm
p_value = norm.cdf(z_score)
print(p_value)
```

### Draw Normal Probability Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
norm_s=np.arange(lower,upper)
plt.plot(norm_s, norm.pdf(norm_s, s_mean, s_sd))
```

### Draw Standard Normal Probability Distribution

```python
standard_data = []
for x in norm_s:
    zs=(x-s_mean)/s_sd
    standard_data.append(zs)
    
import numpy as np

x_all = np.arange(standard_data[0], standard_data[-1], 0.01)
plt.plot(standard_data, norm.pdf(standard_data, st.mean(standard_data), st.stdev(standard_data)), color="green")
```

## Demo - 5.6: Standard Normal Distribution

```python
from scipy.stats import norm
```

```python
'''Consider, 
   the average American monthly household cellular phone bill is USD70.
   Suppose local monthly household cell phone bills are
   normally distributed with a standard deviation of USD11.35.'''
   
'''5.6.1 What is the probability of a randomly selected monthly cell phone bill is
between USD55 and USD80?'''

'''5.6.2. What is the probability that a randomly selected monthly cell phone bill is
no more than USD40?'''
```

```python
x1 = 55
x2 = 80
x3 = 40

mean = 70
sd = 11.35

z1=(x1-mean)/sd
z2=(x2-mean)/sd
z3=(x3-mean)/sd

print("z-score of $55 is =",z1)
print("z-score of $80 is =",z2)
print("z-score of $40 is =",z3)

# print("The probability associated with z="+str(z1) +" is " +str(norm.cdf(z1)))
# print("The probability associated with z="+str(z2) +" is " +str(norm.cdf(z2)))
# print("The probability associated with z="+str(z3) +" is " +str(norm.cdf(z3)))

print("The probability associated with z=",z1," is ",norm.cdf(z1))
print("The probability associated with z=",z2," is ",norm.cdf(z2))
print("The probability associated with z=",z3," is ",norm.cdf(z3))

import matplotlib.pyplot as plt
import numpy as np

data_s=np.arange(30,110)
plt.plot(data_s, norm.pdf(data_s, 70, 11.35))
```
